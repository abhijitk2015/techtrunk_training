```python
img = cv2.imread(r"C:\Python36\images\surprised3.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img = cv2.resize(img,(550,550))
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
def image_zoom(img,zoomsize=2):
    arow,acol,depth = img.shape
    # expanding the image by zoomsize factor
    img = cv2.resize(img,(int(arow*zoomsize),int(acol*zoomsize)))
    # cropping from center
    # get the center of the image
    center = int(img.shape[0]/2),int(img.shape[1]/2)
    # crop the image of the original size
    arow_half = int(arow/2)
    acol_hlaf = int(acol/2)
    img = img[center[0]-arow_half:center[0]+arow_half,center[1]-acol_hlaf:center[1]+acol_hlaf,:]
    return img
    
```

```python
img2 = image_zoom(img,zoomsize=2.8)
plt.imshow(img2)
plt.show()
```

```python
# image rotation
```

```python
# step1 get the rotation matrix
M = cv2.getRotationMatrix2D((img.shape[0]/2,img.shape[1]/2),-20,1)
```

```python
img2 = cv2.warpAffine(img,M,(img.shape[0],img.shape[1]))
img2 = image_zoom(img2,zoomsize=1.5)
noise = numpy.random.randint(0,150,img2.shape)
img2 = img2 +noise
plt.imshow(img2)
plt.show()
```
