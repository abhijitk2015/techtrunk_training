# Morphological operation

```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Python36\images\np2.jpg",0)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
# Erosion
```

```python
kernel = numpy.ones((8,8))
img2 = cv2.erode(img,kernel)
plt.imshow(img2,cmap='gray')
plt.show()
```

```python
#dilation
img3 = cv2.dilate(img,kernel)
plt.imshow(img3,cmap='gray')
plt.show()
```

```python
plt.imshow(img,cmap='gray')
plt.show()
```

```python
img2 = cv2.erode(img,kernel)
img2 = cv2.dilate(img2,kernel)
plt.imshow(img2,cmap='gray')
plt.show()
```
