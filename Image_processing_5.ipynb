```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Python36\images\demo.jpg")
#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
def find_red(img):
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    top_left=[]
    for i in range(0,img.shape[0],50):
        for j in range(0,img.shape[1],50):
            #crop the image upto 50x50
            slice_img = img[i:i+50,j:j+50,:]
            rc=0
            # iterate through every pixel in image and check if it red color
            for m in range(slice_img.shape[0]):
                for n in range(slice_img.shape[1]):
                    if slice_img[m,n,0]>200 and slice_img[m,n,1]<80 and slice_img[m,n,2]<80:
                        rc=rc+1
            # if the red content is more than 5000 pixels out of 10000 then draw it
            if rc>50:
                top_left.append((j,i))
    return top_left
top_left = find_red(img)
for (i,j) in top_left:
    cv2.rectangle(img,(i,j),(i+50,j+50),(0,0,255),5)
plt.figure()
plt.imshow(img)
plt.show()
```

```python
vid = cv2.VideoCapture(0)
while True:
    ret,img = vid.read()
    top_left = find_red(img)
    for (i,j) in top_left:
        cv2.rectangle(img,(i,j),(i+50,j+50),(255,0,0),5)
    cv2.imshow('img',img)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
vid.release()
cv2.destroyAllWindows()
    
```
